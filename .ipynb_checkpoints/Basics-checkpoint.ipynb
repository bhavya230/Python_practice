# String and its functions
String is a class .It is immutable.It is a ordered collection of elements.

```python
# converting whole string to uppercase
"hello world".upper()
```

```python
# converting whole string to lowercase
"Hello World".lower()
```

```python
#converting the string to a title
"hello world".title()
```

```python
#counting occirences of a substr in the string
'hello world'.count('l')
```

```python
#replacing a substring with new substring
'hello world'.replace("hello","hi")
```

```python
#splitting the string into list using a delitmator
"hello world".split('l')
```

```python
#string concatenation
str1="hello"
str2="bye"
#str1.concat(str2) is not a valid method for concatenation in python
str1+" "+str2
```

```python
#formatted strings-f"{}" in curly bracless we can write variables
f"{str1} bhavya. Nice to meet you now {str2}"
```

# LISTS
List are ordered collection of elements. It is like a array but it can store heterogeneous values. 
LIST ARE MUTABLE.

```python
countries=['united states','india','china','brazil']
countries
```

```python
print(countries[0])
```

```python
countries[::-1] # reversed the list
```

```python
#negative indexing
countries[-1] #last element
```

```python
#slicing--extractibg a part of list
#list[start:stop:step size] stop is not included
countries[::2]
```

### Sorting a list

```python
numbers=[2,1,5,3,7,8]
```

```python
numbers.sort() #ascending sorting
numbers
```

```python
numbers.sort(reverse=True) # descending sorting
numbers
```

### Update an Element

```python
numbers=[2,1,5,3,7,8]
```

```python
numbers[0]=10
numbers
```

### Copying a list

```python
countries=['united states','india','china','brazil']
countries
```

```python
# copyig with assignmment operator- does not create a new list.
# only craetes a reference to the same list in the memory
new_list=countries
new_list
```

```python
#copy function- creates a new list (shallow copy)
new_list=countries.copy()
new_list
```

# Dictionary
Dictionary are unordered collection of elements. It is mutable.
Dictionary is the collection of key:value pairs. Dictionary elements can be indexed by key. It is heterogenous in nature.

```python
my_dict={'name':'bhavya','age':20}
my_dict
```

```python
# getting all the keys using .keys()
my_dict.keys()
```

```python
# getting all the values - using .values()
my_dict.values()
```

```python
# getting all dict items key and value -- using .items() 
my_dict.items()
```

```python
# Adding a key value pair
my_dict['height']=5.3
```

```python
my_dict
```

```python
# updating a key value --- using update()
my_dict['height']=5.5
```

```python
my_dict
```
