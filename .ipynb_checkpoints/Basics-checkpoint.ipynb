# String and its functions
String is a class .It is immutable.It is a ordered collection of elements.

```python
# converting whole string to uppercase
"hello world".upper()
```

```python
# converting whole string to lowercase
"Hello World".lower()
```

```python
#converting the string to a title
"hello world".title()
```

```python
#counting occirences of a substr in the string
'hello world'.count('l')
```

```python
#replacing a substring with new substring
'hello world'.replace("hello","hi")
```

```python
#splitting the string into list using a delitmator
"hello world".split('l')
```

```python
#string concatenation
str1="hello"
str2="bye"
#str1.concat(str2) is not a valid method for concatenation in python
str1+" "+str2
```

```python
#formatted strings-f"{}" in curly bracless we can write variables
f"{str1} bhavya. Nice to meet you now {str2}"
```

# LISTS
List are ordered collection of elements. It is like a array but it can store heterogeneous values. 
LIST ARE MUTABLE.

```python
countries=['united states','india','china','brazil']
countries
```

```python
print(countries[0])
```

```python
countries[::-1] # reversed the list
```

```python
#negative indexing
countries[-1] #last element
```

```python
#slicing--extractibg a part of list
#list[start:stop:step size] stop is not included
countries[::2]
```

### Adding elements to a list

```python
countries
```

```python
countries.append('canada') # add element at the end
countries
```

```python
# insert()- adding the element at the particular index
countries.insert(1,'south korea')
```

```python
countries
```

```python
# Adding 2 list
countries_2=['iran','israel']
```

```python
countries+countries_2
```

```python
# nested list
nested_list=[countries,countries_2]
```
