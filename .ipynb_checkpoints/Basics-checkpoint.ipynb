# String and its functions
String is a class .It is immutable.It is a ordered collection of elements.

```python
# converting whole string to uppercase
"hello world".upper()
```

```python
# converting whole string to lowercase
"Hello World".lower()
```

```python
#converting the string to a title
"hello world".title()
```

```python
#counting occirences of a substr in the string
'hello world'.count('l')
```

```python
#replacing a substring with new substring
'hello world'.replace("hello","hi")
```

```python
#splitting the string into list using a delitmator
"hello world".split('l')
```

```python
#string concatenation
str1="hello"
str2="bye"
#str1.concat(str2) is not a valid method for concatenation in python
str1+" "+str2
```

```python
#formatted strings-f"{}" in curly bracless we can write variables
f"{str1} bhavya. Nice to meet you now {str2}"
```

# Dictionary
Dictionary are unordered collection of elements. It is mutable.
Dictionary is the collection of key:value pairs. Dictionary elements can be indexed by key. It is heterogenous in nature.

```python
my_dict={'name':'bhavya','age':20}
my_dict
```

```python
# getting all the keys using .keys()
my_dict.keys()
```

```python
# getting all the values - using .values()
my_dict.values()
```

```python
# getting all dict items key and value -- using .items() 
my_dict.items()
```

```python
# Adding a key value pair
my_dict['height']=5.3
```

```python
my_dict
```

```python
# updating a key value --- using update()
my_dict['height']=5.5
```

```python
my_dict
```

### Copying a dictionary

```python
new_dict=my_dict.copy()
```

```python
new_dict
```

### Remove Element

```python
my_dict.pop('name') #pop method
```

```python
my_dict
```

```python
del my_dict['age']
```

```python
my_dict
```

```python
# clearnig whole dectionary - clear() method
my_dict.clear()
```

```python
my_dict
```

# If-Elif-Else Statements

```python
age= 18

if age>=18:
    print(" you are an adult")
elif age>=13:
    print("you are an teenager")
else:
    print("you are a kid")
```

# For loop

```python
countries
```

```python
# range function
for i in range(5):
    print(i)
```

```python
# iterating iterable list
for country in countries:
    print(country)
```

```python
# iterating using enumerate
for index, country in enumerate(countries):
    print(f"{index} th element is {country}")
```

```python
# itertaing thorugh dictionary
my_dict={'name': 'bhavya', 'age': 20, 'height': 5.5}
my_dict
```

```python
for key,value in my_dict.items(): #prints key and value
    print(f"{key} : {value}")
```

```python
for i in my_dict: #prints only keys
    print(i)
```
