# String and its functions
String is a class .It is immutable.It is a ordered collection of elements.

```python
# converting whole string to uppercase
"hello world".upper()
```

```python
# converting whole string to lowercase
"Hello World".lower()
```

```python
#converting the string to a title
"hello world".title()
```

```python
#counting occirences of a substr in the string
'hello world'.count('l')
```

```python
#replacing a substring with new substring
'hello world'.replace("hello","hi")
```

```python
#splitting the string into list using a delitmator
"hello world".split('l')
```

```python
#string concatenation
str1="hello"
str2="bye"
#str1.concat(str2) is not a valid method for concatenation in python
str1+" "+str2
```

```python
#formatted strings-f"{}" in curly bracless we can write variables
f"{str1} bhavya. Nice to meet you now {str2}"
```
